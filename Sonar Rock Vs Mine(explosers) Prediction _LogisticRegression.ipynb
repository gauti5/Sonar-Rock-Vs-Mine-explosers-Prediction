```python
x=data.drop('R',axis=1)
y=data['R']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
training_data_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(training_data_prediction,y_train)
print("Accuracy Score for the training data is :",training_data_accuracy * 100,'%')
```

```python
test_data_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(test_data_prediction,y_test)
print("Accuracy Score for the test data is :",test_data_accuracy * 100,'%')
```
